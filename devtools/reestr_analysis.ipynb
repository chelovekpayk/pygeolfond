```python
import pandas as pd
import numpy as np

from import_path import imp_path
path = imp_path()
```

Загрузка последней выгрузки из Реестра Роснедр: excel

```python
df = pd.read_excel(path)
df.isnull().sum()
```

```python
df['СубъектРФ'].value_counts(ascending=True, dropna=False)
```

Поиск сведений по номеру лицензии

```python
#Фильтр по одному условию
cond = 'Газпром нефть'
df.loc[df['Наименоваение недропользователя'].str.contains(cond, case=False) == True]
```

```python
#Фильтр по двум условиям
cond = 'Энерком'
df.loc[(df['Наименоваение недропользователя'].str.contains(cond, case=False) == True) & (df['Last'] == True)]
```

```python
#Количество лицензий по субъектам
condition = (df['Last'] == True) & (df['prev_lic'].isna() == True)

country_part_lic = df[condition].groupby("СубъектРФ")['Государственный регистрационный номер'].aggregate('count').sort_values(ascending=False)
country_part_lic[country_part_lic > 100]
```

```python
#Визуализация количества лицензий
country_part_lic[country_part_lic > 100].sort_values(ascending=True).plot.barh()
```
