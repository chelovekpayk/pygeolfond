```python
import pandas as pd
```

```python
df = pd.read_excel(io='../data.xlsx').drop(df.columns[0], axis=1)
```

```python
df.head()
```

```python
#Таблица владельцев лицензии - компания + ИНН
owners = df[['owner','INN']].drop_duplicates(keep='first')
owners['id'] = range(0, len(owners))

#присоединение столбца ID
df = pd.merge(df, owners[['owner', 'id']], on=['owner']).rename(columns={'id':'owner_id'}).drop(columns=['owner'])
```

```python
place = df[['name']].drop_duplicates(keep='first')
place['id'] = range(0, len(place))
```

```python
state = df['state'].drop_duplicates(keep='first')
state['id'] = range(0, len(state))
```
