Извлечение ИНН

```python
#Извлечение ИНН
print('Извлекаю ИНН недропользователей.')
pattern_for_inn = '([\d{10}|\d{12}]+)'
pattern_for_replace_inn = '(\(ИНН.*\)$)'

df['INN'] = df['Сведения о пользователе недр'].str.extract(pattern_for_inn)
df['Наименоваение недропользователя'] = df['Сведения о пользователе недр'].replace(pattern_for_replace_inn, '', regex=True).str.strip()
df[['INN', 'Сведения о пользователе недр', 'Наименоваение недропользователя']].head()
```

```python
#Проверки
df['INN'][(df['INN'].isna() == True) & (df['Сведения о пользователе недр'].isna() == False)] #Количество пустых записей ИНН
df['Сведения о пользователе недр'][df['Сведения о пользователе недр'].str.contains(pattern_for_replace_inn, regex=True) == False] #Количество Сведений без ИНН
```

Извлечение данных по лицензиям

```python
#from dataclasses import replace

df['Last'] = np.where(df['Сведения о переоформлении лицензии на пользование недрами'].isna(), True, False)


pattern_for_lic = '([А-Я]+[0-9]+[-А-Я]+)' #Паттерн для извлечения номера ранее выданной лицензии
pattern_for_date = '([\d\.]+$)' #Паттерн для извлечения даты ранее выданной лицензии

#Извлечение данных предыдущих лицензий
df['prev_lic'] = df['Ранее выданные лицензии'].str.extract(pattern_for_lic)
df['prev_date'] = df['Ранее выданные лицензии'].str.extract(pattern_for_date)

#Извлечение данных будущих лицензий
df['forw_lic'] = df['Сведения о переоформлении лицензии на пользование недрами'].str.extract(pattern_for_lic)
df['forw_date'] = df['Сведения о переоформлении лицензии на пользование недрами'].str.extract(pattern_for_date) 

#Преобразование столбца в датавремя и добавление 1 года для совпадения с ГБЗ`; Извлечение года лицензии
df['prev_date'] = pd.to_datetime(df['prev_date'], format='%d.%m.%Y', errors='ignore', dayfirst=True) + pd.offsets.DateOffset(years=1) 
df['prev_date'] = df['prev_date'].dt.year.astype(str).str.replace('.0', '', regex=False) 

df['forw_date'] = pd.to_datetime(df['forw_date'], format='%d.%m.%Y', errors='ignore', dayfirst=True) + pd.offsets.DateOffset(years=1) 
df['forw_date'] = df['forw_date'].dt.year.astype(str).str.replace('.0', '', regex=False)
```
