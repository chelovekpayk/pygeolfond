```python
import pandas as pd
import numpy as np

import os
import datetime
```

```python
#Загрузка файла
#Даты добавлены в app

fpath = os.path.normpath(os.path.expanduser('~/Dropbox/Проекты/Роснедра/Реестр/reestr_request_last.csv'))

cols = [
    'Государственный регистрационный номер',
    'Дата лицензии',
    'СубъектРФ','Вид полезного ископаемого','Наименование участка недр',
    'INN','Наименоваение недропользователя',
    'Last','Year',
    'N', 'E',
    'prev_lic', 'prev_date',
    'forw_lic', 'forw_date',
    ]
types = {
    'Государственный регистрационный номер':'str',
   # 'Дата лицензии':'datetime64[ns]',
    'Last':'bool',
    'N':'str', 'E':'str',
    'prev_lic':'str', #'prev_date':'datetime64[ns]',
    'forw_lic':'str', #'forw_date':'datetime64[ns]',
    'Наименование участка недр':'str', 'Вид полезного ископаемого':'str',
    'СубъектРФ':'str', 'INN':'str', 'Year':'int',
    }
dates = ['Дата лицензии','prev_date','forw_date' ]

df = pd.read_csv(fpath, sep=',', usecols=cols, dtype=types, parse_dates=dates).set_index('Государственный регистрационный номер')
```

Извлечение данных о предыдущих владельцах лицензий.

    Скрипт извлекает столбец "Наименование недропользователя" и присоединяет столбец к датафрейму справа 

```python
#Блок извлечения предыдуших владельцев
#Добавлено в app.py
prev_owner_df = df.reset_index().set_index('forw_lic')
prev_owner_df = prev_owner_df[prev_owner_df.index.notnull()]
prev_owner_df = prev_owner_df[~prev_owner_df.index.duplicated(keep='last')]['Наименоваение недропользователя'].rename('prev_owner')
```

```python
#Блок присоединения данных
df2 = pd.concat([df,prev_owner_df], axis=1, join='outer').rename_axis('Государственный регистрационный номер')
df2.reset_index(inplace=True)
df2 = df2[df2['Дата лицензии'].notna()]
```
